#### 构建movie数据源

```python
import pandas as pd

# 读取 file
df1 = pd.read_csv("movie_data_en.csv")
df2 = pd.read_csv("movie_data_cn.csv")
```

```python
df1
```

```python
df2
```

```python
# 纵向合并
df = pd.concat([df1, df2], axis=0)

# 如不需要保留原索引，重置索引
df = df.reset_index(drop=True)

df
```

```python
df_del1row = df.iloc[:, 1:]

# 对行随机洗牌
df_shuffled = df_del1row.sample(frac=1).reset_index(drop=True)
df_shuffled
```

```python
# 生成与行索引相同的字符串列
df_shuffled.insert(
    0,                      # 插入到最前面
    "ID",            # 列名
    df_shuffled.index.astype(str)
)
```

```python
df_shuffled
```

```python
print(df_shuffled)
```

```python
import csv

df_shuffled.to_csv(
    "movies.csv",
    index=False,
    quoting=csv.QUOTE_ALL
)
```
